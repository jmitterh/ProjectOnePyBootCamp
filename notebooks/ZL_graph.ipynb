```python
#imports 
import pandas as pd
import numpy as np
```

```python
# cleaned data file to be saved
# cleanData = "../ZL/master_clean_data.csv"
cleanData = "../data/cleanData/master_clean_data.csv"

# path csv file location
pathCleanData ="../data/master_clean_data/"
```

```python
#calling the csv raw data
cleansed_data = pd.read_csv(cleanData,index_col=0)
cleansed_data.head()
```

```python
# date = pd.DataFrame(cleansed_data['date created'])
dates = pd.to_datetime(cleansed_data['date created'])
cleansed_data['year'] = dates.map(lambda x:x.year)
cleansed_data['month'] = dates.map(lambda x:x.month)
cleansed_data.head()
```

```python
g = cleansed_data.groupby('category')['month'].value_counts()
plotable_1 = g.unstack(level='category').fillna(0)
plotable_1
```

```python
top_5c = cleansed_data.groupby('category').count()['month'].sort_values(ascending=False).reset_index().head()

plotable_1 = plotable_1[list(top_5c['category'])]

category = plotable_1.plot(kind='line',legend = True,title = 'Top 5 Job Categories',xticks=plotable_1.index).legend(bbox_to_anchor=(2, 1))     
```

```python
g_2 = cleansed_data.groupby(['state'])['month'].value_counts()
plotable_2 = g_2.unstack(level='month').fillna(0)

top_5 = cleansed_data.groupby('state').count()['month'].sort_values(ascending=False).reset_index().head()

plotable_2 = plotable_2.T[list(top_5['state'])]
plotable_2
```

```python
plotable_2.plot(title = 'Top 5 States',xticks=plotable_2.index).legend(bbox_to_anchor=(1.5, 1))
```

```python
g_3 = cleansed_data.groupby(['company name'])['month'].value_counts()
plotable_3 =  g_3.unstack(level='month').fillna(0)

top_5e = cleansed_data.groupby('company name').count()['month'].sort_values(ascending=False).reset_index().head()

plotable_3 = plotable_3.T[list(top_5e['company name'])]
plotable_3
```

```python
plotable_3.plot( title = 'Top 5 Employers',xticks=plotable_3.index).legend(bbox_to_anchor=(1.5, 1))
```
